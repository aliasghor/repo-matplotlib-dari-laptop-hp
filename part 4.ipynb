```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')
```

```python
mu, sigma = 100, 15 # Nilai Mean Dan Nilai Standart Deviatition

x = mu + sigma * np.random.randn(10000)# Normal Distribution
x
```

```python
x.shape
```

```python
plt.hist(x, bins=50, facecolor='g', alpha=0.75)

plt.title('Contoh Histogram')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.text(47, 550, r'$\mu=100,\ \sigma=15$')

plt.tight_layout()
plt.grid()
plt.show()
```

# Histogram Menggunakan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x, bins=50, facecolor='r')

ax.set_title('Contoh Historgram')
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

ax.text(45, 550, '$\mu=100,\ \sigma=15$')
ax.grid()
plt.show()
```
