```python
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python
data = [2000,1500,4500,5500,6000]
kategori = ['C','C++','JavaScript','Java','Python']
colors = ['blue','cyan','yellow','red','green']
explode = [0, 0, 0.2, 0, 0.1]
```

```python
plt.pie(data, labels=kategori, autopct='%1.1f%%', startangle=90)

plt.title('Simple Pie Chart')

plt.show()
```

```python
plt.pie(data, labels=kategori, colors=colors, autopct='%1.1f%%')

plt.title('Simple Pie Chart')

plt.legend()
plt.show()
```

```python
plt.pie(data, labels=kategori, colors=colors, autopct='%1.1f%%', explode=explode)

plt.title('Simple Pie Chart')

plt.legend()
plt.show()
```
