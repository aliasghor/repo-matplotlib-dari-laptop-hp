```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = np.linspace(1,10,15)
x
```

```python
plt.plot(x,x,'bs-',label='lin')
plt.plot(x,np.log(x),'ro--',label='log')

plt.title('Contoh Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.legend()
plt.tight_layout()
plt.show()
```

# Multiple Style

```python
plt.style.use(['seaborn','dark_background'])
```

```python
fig, ax = plt.subplots()

ax.plot(x,x,marker='s',linestyle='-',label='lin')
ax.plot(x,np.log(x),marker='o',linestyle='--',label='log')

ax.set_title('Contoh Plot')
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

ax.legend()
plt.show()
```

# Temporary Styling

```python
plt.style.use('ggplot')
```

```python
plt.plot(np.sin(np.linspace(0,2 * np.pi)), 'ro-')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0,2 * np.pi)), 'ro-')
    
plt.show()
```
