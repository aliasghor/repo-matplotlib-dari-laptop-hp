```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

print(matplotlib.__version__)
print(np.__version__)

plt.style.use('dark_background')
```

```python
x = [10,20,30,40,50]
y = [1000,2000,3000,4000,5000]

plt.plot(x, y, label='Salaries', marker='o')

plt.title('Salaries Based On Ages')
plt.xlabel('Ages')
plt.ylabel('Salaries')

plt.show()
```

```python
y = [2,5,7,9,10]

plt.plot(y, c='r', marker='o')

plt.ylabel('Sumbu Y')

plt.show()
```

```python
x = [1,2,3,4]
y = [5,7,4,12]

plt.plot(x, y, 'b-')
plt.axis([0, 5, 0, 15])

plt.show()
```

```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, label='linear', linestyle='--')
plt.plot(t, t**2, label='Quadratic', marker='s')
plt.plot(t, t**3, label='Cubic', marker='^')

plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')

plt.legend()
plt.show()
```
