# Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Membuat Plotting Sederhana

```python
plt.plot([2,5,7,11,14])
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')
plt.show()
```

```python
plt.plot([2,5,7,11,14],[14,11,7,5,2])
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'g^')

plt.axis([0,6,0,20])

plt.show()
```

# Multi Plot Dalem Satu Pemanggilan fungsi plot()

```python
x = np.arange(0,5,0.2)
x
```

```python
plt.plot(x,x,'r--')
plt.plot(x,x**2,'go')
plt.plot(x,x**3,'bs')

plt.show()
```

# Kita Juga Dapat Melakukan Multi Plot Dalam Satu Pemanggilan Fungsi plot()

```python
plt.plot(x,x,'ro',
         x,x**2,'b^',
         x,x**3,'gs')

plt.show()
```

# Membuat Sebuah Data Frame

```python
data = {
    "a":np.arange(50),
    "c":np.random.randint(0,50,50),
    "d":np.random.randn(50)
}

df = pd.DataFrame(data)

df['b'] = df['a'] + 10 * np.random.randn(50)
df['d'] = np.abs(df['d']) * 100

df
```

```python
data = {
    "a":np.arange(50),
    "c":np.random.randint(0,50,50),
    "d":np.random.randn(50)
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a','b',data=df)

plt.xlabel('Entry a')
plt.ylabel('Entry b')

plt.show()
```

```python
plt.scatter(x='a',y='b',c='c',s='d',data=data)

plt.xlabel('Entry a')
plt.ylabel('Entry b')

plt.show()
```
