```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

# Simple Violin plot

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Pengaturan Pada Violin Plot

```python
plt.violinplot(data,showextrema=True,showmeans=True,showmedians=True,quantiles=[0.25,0.5,0.75])

plt.title('Pengaturan Pada Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Horizontal Violin Plot

```python
plt.violinplot(data,vert=False,showextrema=True,showmeans=True,showmedians=True,quantiles=[0.25,0.5,0.75])

plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Multple Violin Plot

```python
np.random.seed(10)
data1 = np.random.normal(100,10,200)
data2 = np.random.normal(90,30,200)
data3 = np.random.normal(80,25,200)
data4 = np.random.normal(70,20,200)
# positions = [1, 2, 3, 4]

data = [data1,data2,data3,data4]
```

```python
plt.violinplot(data,showextrema=True,showmeans=True,showmedians=True)

plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```
