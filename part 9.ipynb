```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
```

```python
def gambar():
    data1 = [25,85,75,40,60]
    data2 = [40,35,20,55,10]
    kategori = ['A','B','C','D','E']
    
    x = np.arange(len(kategori))
    
    width = 0.35
    
    plt.bar(x-width/2,data1,width,label='Data 1')
    plt.bar(x+width/2,data2,width,label='Data 2')
    
    plt.xticks(x,kategori)
    
    plt.grid(linestyle='--',axis='y',linewidth=1,alpha=0.75)
    
    plt.title('Contoh Groupped Bar Plot')
    plt.xlabel('Sumbu X')
    plt.ylabel('Sumbu Y')
    
    plt.legend()
    plt.show()
```

```python
gambar()
```
