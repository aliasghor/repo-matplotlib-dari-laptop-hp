```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'A':1000,
    'B':2000,
    'C':3000,
    'D':4000,
    'E':5000,
    'F':6000,
    'G':7000,
    'I':8000,
    'J':9000
}
data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

ax.grid()
plt.show()
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots()

ax.bar(items,count)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Simple Bar Vertikal plot')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

labels = ax.get_xticklabels()

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Simple Bar Horizontal plot')

plt.setp(labels,rotation=45
         ,horizontalalignment='right')

fig.tight_layout()
plt.show()
```

# Pengaturan Format Pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x,pos):
    return f"{int(x/1000)}K"
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Plot')

fig.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

ax.axvline(np.mean(count),linestyle='-',c='r')

ax.set_title('Horizontal Bar Plot')
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)

ax.axvline(2000,linestyle='--')

ax.set_title('Horizontal Bar Plot')
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

fig.tight_layout()
plt.show()
```
