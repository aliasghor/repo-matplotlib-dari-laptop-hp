```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

print(matplotlib.__version__)
print(np.__version__)
```

```python
np.random.seed(15)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

# Simple Box Plot

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Notched Box Plot

```python
plt.boxplot(data,labels=['Data'],notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Pengaturan Marker Untuk Outlier

```python
plt.boxplot(data,labels=['Data'],notch=True,showfliers=True,flierprops={'markerfacecolor':'r','marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Horizontal Box Plot

```python
plt.boxplot(data,labels=['Data'],vert=False)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Multiple Box Plot

```python
np.random.seed(10)
data1 = np.random.normal(loc=100,scale=10,size=200)
data2 = np.random.normal(loc=80,scale=30,size=200)
data3 = np.random.normal(loc=90,scale=20,size=200)
data4 = np.random.normal(loc=70,scale=25,size=200)

data = [data1,data2,data3,data4]
labels = ['Data 1','Data 2','Data 3','Data 4']
```

```python
plt.boxplot(data,labels=labels,notch=True,showfliers=True,flierprops={'markerfacecolor':'b','marker':'s'})

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```
