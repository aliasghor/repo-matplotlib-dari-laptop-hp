```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')
```

```python
x = [1,2,3,4]
y = [3,2,1,4]

fig , ax = plt.subplots()

ax.plot(x, y, marker='o')

plt.title('Mengulas Kembali Materinya')

plt.show()
```

```python
x = [20,21,22,25,28,29,30]
y = [600,800,1000,1200,1350,1500,2000]

plt.plot(x, y, linestyle='--')

plt.title('Nyoba Doang')
plt.xlabel('Ages')
plt.ylabel('Salaries')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 3) # A Figure That Has 2 Columns And 3 Rows
```

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots()

ax.plot(x, x, label='Linear')
ax.plot(x, x**2, label='Quadratic')
ax.plot(x, x**3, label='Qubic')

ax.set_title('Simple Plot')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

ax.legend()
plt.tight_layout()
plt.show()
```
