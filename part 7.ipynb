```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_uas = [
    ["Gerry",100],
    ["Sampi",85],
    ["Pace",75],
    ["Mogi",95],
    ["Balu",87]
]
```

```python
data_uas
```

# Table Plot Menggunakan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas, loc='center')

table.set_fontsize(18)
table.scale(1.5, 4) # Ukuran Kolom Baris

ax.axis(False)

plt.show()
```

# Table Plot Menggunakan Pyplot Style

```python
table = plt.table(cellText=data_uas, loc='center')

table.set_fontsize(18)
table.scale(1.5, 5)

ax = plt.gca()
ax.axis(False)

plt.show()
```
