```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
import numpy as np

print(matplotlib.__version__)
print(pd.__version__)
```

```python
data = {
    "tahun":["1958","1963","1968","1973","1978","1983","1988","1993","1998","2003","2008","2013","2018"],
    "populasi":[51652500,53624900,55213500,56223000,56178000,56223000,56178000,56315000,56916000,57713000,58474000
               ,59636000,61823000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.plot(x='tahun',y='populasi',marker='s',c='r',linestyle='--',kind='line')

plt.title('Line Plot')
plt.ylabel('Jumlah Populasi')

plt.grid()
plt.show()
```

```python
devs = {
    "Devs":["Python","JavaScript","C++","Java","C"],
    "Salaries":[6500,5450,5000,4780,3500]
}
```

```python
df = pd.DataFrame(devs,columns=['Devs','Salaries'])
df
```

```python
df.plot(kind='bar',x='Devs',y='Salaries')

plt.title('Developer Salaries')
plt.ylabel('Developer')

plt.show()
```

```python
df.plot(kind='barh',x='Devs',y='Salaries')

plt.title('Developer Salaries')
plt.ylabel('Developer')

plt.show()
```

```python
df.plot(kind='barh',x='Devs',y='Salaries',color='r',legend=False)

plt.title('Developer Salaries')
plt.ylabel('Developer')

plt.show()
```

```python
devs = {
    "Devs":["Python","JavaScript","C++","Java","C"],
    "Salaries":[6500,5450,5000,4780,3500],
    "BonusIncome":[500,450,400,300,250]
}
```

```python
df = pd.DataFrame(devs,columns=['Devs','Salaries','BonusIncome'])
df
```

```python
df.plot(kind='scatter',x='Salaries',y='BonusIncome',color='cyan')

plt.title('Scatter Plot')
plt.ylabel('Bonus Income')

plt.show()
```

```python
devs = {
    "Devs":["Python","JavaScript","C++","Java","C"],
    "Salaries":[6500,5450,5000,4780,3500]
}
```

```python
df = pd.DataFrame(devs)
df
```

```python
plt.pie(df['Salaries'],labels=df['Devs'],autopct='%1.1f%%',explode=[0.1,0,0,0,0])

plt.title('Pie Chart')

plt.show()
```

```python
df.plot(kind='pie',y='Salaries',figsize=(6,6))

plt.title('Pie Chart')

plt.show()
```

```python
df = df.set_index('Devs')
df
```

```python
df.plot(kind='pie',y='Salaries',figsize=(6,6))

plt.title('Pie Chart')

plt.show()
```
