```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# plt.style.use('seaborn-v0_8-whitegrid')
print(matplotlib.__version__)
print(np.__version__)
```

```python
data = [25, 45, 55, 125, 225]
kategori =['A', 'B', 'C', 'D', 'E',]
```

```python
plt.bar(kategori, data)

plt.title('Bar Plot Example')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.tight_layout()
plt.show()
```

```python
data = [250, 450, 550, 1250, 2000]
kategori =['A', 'B', 'C', 'D', 'E',]
```

```python
plt.bar(kategori, data, color='red', alpha=0.75)

plt.grid(linestyle='--', linewidth=2, axis='y', alpha=1)

plt.title('Bar Plot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.tight_layout()
plt.show()
```

```python
data = [250, 450, 550, 1250, 2000]
kategori =['A', 'B', 'C', 'D', 'E',]
data.reverse()
kategori.reverse()

plt.barh(kategori, data)

plt.title('Horizontal Bar Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.tight_layout()
plt.show()
```

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori =['A','B','C','D','E']
```

```python
plt.bar(kategori, data1, label='Data 1')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.title('Stacked Bar Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.legend()
plt.grid(linestyle='--', linewidth=2, axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.bar(kategori, data1, label='Data 1', bottom=data2)
plt.bar(kategori, data2, label='Data 2')

plt.title('Stacked Bar Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.legend()
plt.grid(linestyle='--', linewidth=2, axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori =['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x - width/2, data1, width, label='Data 1')
plt.bar(x + width/2, data2, width, label='Data 2')

plt.title('Groupped Bar Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.legend()
plt.grid(linestyle='--', linewidth=3, axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
