```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
devs = ['Python','JavaScript','C++','Java']
salaries = [6500,5000,4650,4000]
```

# Simple 2x2 Layout With Subplots

```python
fig, ax = plt.subplots(ncols=2,nrows=2,constrained_layout=True)

ax[0][0].bar(devs,salaries)
ax[0][1].pie(salaries,labels=devs,explode=[0.1,0,0,0],autopct='%1.1f%%')
ax[1][0].pie(salaries,labels=devs,explode=[0,0.1,0,0],autopct='%1.1f%%')
ax[1][1].barh(devs,salaries)

plt.show()
```

# Simple 2x2 Layout With GridSpec

```python
fig = plt.figure(constrained_layout=True)
gs = gridspec.GridSpec(ncols=2,nrows=2,figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(devs,salaries)
ax2.pie(salaries,labels=devs,explode=[0.1,0,0,0],autopct='%1.1f%%')
ax3.pie(salaries,labels=devs,explode=[0,0,0.1,0],autopct='%1.1f%%')
ax4.barh(devs,salaries)

plt.show()
```

# Another Way To Plot 2x2 Layout With Gridspec

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=2,nrows=2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(devs,salaries)
ax2.pie(salaries,labels=devs,autopct='%1.1f%%',explode=[0.1,0,0,0])
ax3.pie(salaries,labels=devs,autopct='%1.1f%%',explode=[0,0.1,0,0])
ax4.barh(devs,salaries)

plt.show()
```

# Rows Span And Cols Span With GridSpec

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=2,nrows=2)

ax1 = fig.add_subplot(gs[0,:])
# ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(devs,salaries)
ax2.pie(salaries,labels=devs,autopct='%1.1f%%',explode=[0.1,0,0,0])
ax3.pie(salaries,labels=devs,autopct='%1.1f%%',explode=[0,0.1,0,0])
ax4.barh(devs,salaries)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=3,nrows=3)

ax1 = fig.add_subplot(gs[0,:])
ax1.set_title('gs[0,:]')

ax2 = fig.add_subplot(gs[1,:-1])
ax2.set_title('gs[1,:-1]')
ax2.barh(devs,salaries)

ax3 = fig.add_subplot(gs[1:,-1])
ax3.set_title('gs[1:,-1]')
ax3.bar(devs,salaries)

ax4 = fig.add_subplot(gs[-1,0])
ax4.set_title('gs[-1,0]')

ax5 = fig.add_subplot(gs[-1,-2])
ax5.set_title('gs[-1,-2]')

plt.show()
```
