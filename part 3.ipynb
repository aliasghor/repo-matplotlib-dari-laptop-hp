```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = np.arange(0.0, 2.0 ,0.1)
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots()

ax.plot(x, s)

ax.set_title('Visualisasi Nilai Sin')
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Sin(X)')

plt.tight_layout()
plt.show()
```

```python
plt.plot(x, s)

plt.title('Visualisasi Nilai Sin')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Sin(X)')

plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, c='r', linestyle='--')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2, y2, c='g', marker='.')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai X')

plt.tight_layout()
plt.show()
```

# Multiple Subplots Menggunakan Pyplot style

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1, c='r')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, c='g')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')


plt.tight_layout()
plt.show()
```
