```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.rename(columns={
    "Country name":"Country",
    "Perceptions of corruption":"Corruption",
    "Freedom to make life choices":"Freedom"
},inplace=True)
df.columns
```

# Menggunakan Bar Plot

```python
df[:5].plot(x='Country',y=['Corruption','Freedom','Generosity','Social support'],kind='bar')

plt.title('Contoh Bar Plot')
plt.ylabel('Nilai')

plt.show()
```

# Menggunakan Line Plot

```python
df[:5].plot(x='Country',y=['Corruption','Freedom','Generosity','Social support'],kind='line')

plt.title('Contoh Line Plot')
plt.ylabel('Nilai')

plt.show()
```

# Menggunakan Box Plot

```python
df.plot(y=['Corruption','Freedom','Generosity','Social support'],kind='box')

plt.title('Contoh Box Plot')
plt.xlabel('Parameter')
plt.ylabel('Nilai')

plt.show()
```

# Menggunakan Scatter Plot

```python
df.plot(x='Corruption',y='Freedom',kind='scatter')

plt.title('Scatter Plot')
plt.xlim((0,1))
plt.ylim((0,1))

plt.show()
```

# Menggunakan Histogram

```python
df.plot(x='Country',y='Corruption',bins=20,kind='hist')

plt.title('Histogram Plot')
plt.xlabel('Nilai')

plt.show()
```

```python
df.plot(x='Country',
        y=['Corruption','Freedom','Generosity','Social support'],kind='hist',subplots=True,layout=(2,2))

plt.suptitle('Contoh Histogram Subplots')
plt.tight_layout()

plt.show()
```
