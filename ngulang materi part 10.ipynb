```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(10)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

# Simple Box Plot

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Box Plot With Notch Actived(notch=True)

```python
plt.boxplot(data,labels=['Data'],notch=True)

plt.title('Box Plot With Notch True')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Cutomizing Box Plot With Some Colors

```python
plt.boxplot(data,labels=['Data'],notch=True,showfliers=True,flierprops={'markerfacecolor':'b','marker':'d'})

plt.title('Cutomizing Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Multiple Box Plot

```python
np.random.seed(10)
data1 = np.random.normal(loc=100,scale=10,size=200)
data2 = np.random.normal(loc=90,scale=30,size=200)
data3 = np.random.normal(loc=80,scale=25,size=200)
data4 = np.random.normal(loc=70,scale=25,size=200)

data = [data1,data2,data3,data4]
labels = ['Data 1','Data 2','Data 3','Data 4']
```

```python
plt.boxplot(data,labels=labels,notch=True,showfliers=True,flierprops={'markerfacecolor':'b','marker':'o'})

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```
