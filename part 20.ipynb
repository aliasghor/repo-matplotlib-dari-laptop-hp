```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Sample Dataset

```python
df = pd.DataFrame({
    "nama":["Cecep","Karti","Bejo","Tejo","Asep","Wati","Amat"],
    "usia":[23,78,22,19,45,33,20],
    "JenisKelamin":["L","P","L","L","L","P","L"],
    "Provinsi":["Jabar","Jatim","Jabar","Jatim","Jabar","Jateng","Jateng"],
    "n_anak":[2,0,0,3,2,1,4],
    "n_pets":[5,1,0,5,2,2,3]
})

df
```

# Scatter Plotting

```python
df.plot(kind='scatter',x='n_anak',y='n_pets',c='r')

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Hewan Peliharaan')

plt.show()
```

# Histogram Plotting

```python
df[['usia']].plot(kind='hist',bins=[0,20,40,60,80,100])

plt.title('Contoh Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')

plt.show()
```

# Line Plot

```python
df.plot(kind='line',x='nama',y='n_anak',marker='s',linestyle='--',color='green')

plt.title('Line Plot')
plt.ylabel('Jumlah Hewan')

plt.legend()
plt.grid()
plt.show()
```

```python
df.plot(kind='line',x='nama',y='n_anak',marker='o',linestyle='-',color='y')
df.plot(kind='line',x='nama',y='n_pets',marker='^',linestyle='--',color='r')

plt.title('Line Plot')
plt.ylabel('Jumlah')

plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line',x='nama',y='n_anak',ax=ax,marker='s',linestyle='-',c='y')
df.plot(kind='line',x='nama',y='n_pets',ax=ax,marker='^',linestyle='--',c='r')

plt.title('Multiple Line Plot')
plt.ylabel('Jumlah')

plt.grid()
plt.show()
```

```python
df.plot(kind='bar',x='nama',y='usia')

plt.title('Contoh Bar Plot')
plt.ylabel('Ages')

plt.show()
```

```python
df.groupby('Provinsi')['nama'].count().plot(kind='bar')

plt.title('Contoh Groupped Bar Plot')
plt.ylabel('Jumlah')

plt.show()
```

# Stacked Bar Plot Dengan Dua Level Group By

```python
df.groupby(['Provinsi','JenisKelamin']).size().unstack().plot(kind='bar',stacked=True)

plt.title('Contoh Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
df.groupby(['JenisKelamin','Provinsi']).size().unstack().plot(kind='bar',stacked=True)

plt.title('Stacked Bar Plot')
plt.ylabel('Jumlah')

plt.show()
```
